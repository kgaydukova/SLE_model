```python
# Assuming you have your DataFrame df
corr_matrix, pval_matrix = spearmanr(df)

# Assuming your DataFrame has column names for better interpretation
columns = df.columns

# Print the correlation matrix for better visualization
print(pd.DataFrame(corr_matrix, index=columns, columns=columns))

# Find strongly correlated variables
threshold = 0.7
strongly_correlated_pairs = [(columns[i], columns[j]) for i in range(len(columns)) for j in range(i+1, len(columns)) if abs(corr_matrix[i, j]) > threshold]

print("Strongly correlated variables:")
for var1, var2 in strongly_correlated_pairs:
    print(f"{var1} and {var2}: {corr_matrix[columns.index(var1), columns.index(var2)]}")
```

```python
columns_for_new_df = ['Age at diagnosis', 'SLEDAI-2K', 'Er (in debut)', 'ANF (currently)', 'PRE (currently)']
df = df[columns_for_new_df]
```

```python
df.loc[:, 'Class'] = df['PRE (currently)'].apply(lambda x: 0 if x <= 7.5 else 1)
```

```python
df = df.drop(columns = 'PRE (currently)')
```

```python
df['ANF (currently)'] = df['ANF (currently)'] / 160
df.loc[:, 'ANF (currently)'] = df['ANF (currently)'].apply(lambda x: 0 if x <= 640 else 1)
```

```python
# Assuming you have your DataFrame df with columns 'Class', 'Var1', 'Var2', ... representing the class and numeric variables
# Split the DataFrame into two groups based on 'Class'
group0 = df[df['Class'] == 0]
group1 = df[df['Class'] == 1]

# Perform the t-test for each numeric variable
for column in df.columns:
    if column != 'Class':  # Exclude the 'Class' column from testing
        var0 = group0[column]
        var1 = group1[column]
        
        # Perform the t-test
        t_stat, p_value = stats.ttest_ind(var0, var1)
        
        # Print the results
        print(f"T-test results for '{column}':")
        print(f"T-statistic: {t_stat}")
        print(f"P-value: {p_value}")
        print("---------------------------------")
```

```python
# Assuming you have your DataFrame df with columns 'Class', 'Var1', 'Var2', ... representing the class and numeric variables
# Split the DataFrame into two groups based on 'Class'
group0 = df[df['Class'] == 0]
group1 = df[df['Class'] == 1]

# Create empty lists to store t-test results
columns = df.columns.drop('Class')
t_stats = []
p_values = []

# Perform the t-test for each numeric variable and store results
for column in columns:
    var0 = group0[column]
    var1 = group1[column]
    
    # Perform the t-test
    t_stat, p_value = stats.ttest_ind(var0, var1)
    
    # Append the results to the lists
    t_stats.append(t_stat)
    p_values.append(p_value)

# Adjust the p-values using the chosen method (e.g., Benjamini-Hochberg)
adjusted_p_values = sm.multipletests(p_values, method='fdr_bh')[1]

# Create a DataFrame to display the results
results_df = pd.DataFrame({
    'Variable': columns,
    'T-Statistic': t_stats,
    'P-Value': p_values,
    'Adjusted P-Value': adjusted_p_values
})

print(results_df)
```
